# Запуск тестового модуля

Этот notebook запускает модуль `tests.test_checker`, который проверяет бизнес-логику `ifc_checker_script.check_ifc(...)` на sample-данных проекта.

```python
from pathlib import Path
import subprocess
import sys

repo_root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
print(f'Python executable: {sys.executable}')
repo_root
```

## Запуск `unittest`

Тест сохраняет артефакты в `.output/test_reports` в корне репозитория.

```python
result = subprocess.run(
    [sys.executable, '-m', 'unittest', 'tests.test_checker'],
    cwd=repo_root,
    capture_output=True,
    text=True,
)

print(result.stdout)
print(result.stderr)
print(f'Return code: {result.returncode}')
```

## Полученные артефакты

После выполнения можно посмотреть файлы, которые тест сохранил в `.output`.

```python
sorted(str(path.relative_to(repo_root)) for path in (repo_root / '.output').rglob('*'))
```
